## Overview

The figures in this chapter represent the movement of two populations ($\mathcal{P}^\mu$ and $\mathcal{P}^\nu$) between two sections of one city (sections 1 and 2).  The two populations have internal agglomeration trends Îº (the tendency for a population to want to live with itself) and segregation trends (the tendency for a population to not want to live with the other population), which can be either symmetric or asymmetric.  The figures below explore a range of possibilities by varying the agglomeration and segregation trends.

### Figure 4.1

In this figure, we have a weak internal agglomeration trend and a weak reciprocal segregation trend.  All fluxlines approach the origin, so the homogeneous mixture of populations $\mathcal{P}^\mu$ and $\mathcal{P}^\nu$ is stable.

```python
N = 100                                
x_start, x_end = -1.0, 1.0
y_start, y_end = -1.0, 1.0
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X, Y = numpy.meshgrid(x, y)

Îº_Î¼Î¼ = 0.5
Îº_Î½Î½ = 0.5
Îº_Î¼Î½ = -0.2
Îº_Î½Î¼ = -0.2

dx = (Îº_Î¼Î¼ - 1) * X + Îº_Î¼Î½ * Y
dy = Îº_Î½Î¼ * X + (Îº_Î½Î½ - 1) * X

%matplotlib inline
pyplot.figure(figsize=(10,10))
pyplot.title('Figure 4.1 (a)')
pyplot.xlabel('x(Ï)', fontsize = 14)
pyplot.ylabel('y(Ï)', fontsize = 14)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, dx, dy, linewidth = 1, density = 2)
```

### Figure 4.2

In this figure, we have a moderate internal agglomeration trend and a strong reciprocal segregation trend.  All fluxlines approach one of two equilibrium points in the second and fourth quadrants, so the homogeneous mixture of populations $\mathcal{P}^\mu$ and $\mathcal{P}^\nu$ is unstable, and the population will self-segregate into two ghettos.

```python
N = 100                                
x_start, x_end = -1.0, 1.0
y_start, y_end = -1.0, 1.0
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X, Y = numpy.meshgrid(x, y)

Îº_Î¼Î¼ = 0.5
Îº_Î½Î½ = 0.5
Îº_Î¼Î½ = 1
Îº_Î½Î¼ = 1

dx = (Îº_Î¼Î¼ - 1) * X + Îº_Î¼Î½ * Y
dy = Îº_Î½Î¼ * X + (Îº_Î½Î½ - 1) * X

%matplotlib inline
pyplot.figure(figsize=(10,10))
pyplot.title('Figure 4.2 (a)')
pyplot.xlabel('x(Ï)', fontsize = 14)
pyplot.ylabel('y(Ï)', fontsize = 14)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, dx, dy, linewidth = 1, density = 2)
```

### Figure 4.3

In this figure, we have a moder internal agglomeration trend and a strong asymmetric interaction.  All fluxlines spiral into the center point, so the homogeneous mixture of populations $\mathcal{P}^\mu$ and $\mathcal{P}^\nu$ is stable, and the population will slowly become homogeneous.

```python
N = 100                                
x_start, x_end = -1.0, 1.0
y_start, y_end = -1.0, 1.0
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X, Y = numpy.meshgrid(x, y)

Îº_Î¼Î¼ = 0.5
Îº_Î½Î½ = 0.5
Îº_Î¼Î½ = -1
Îº_Î½Î¼ = 1

dx = (Îº_Î¼Î¼ - 1) * X + Îº_Î¼Î½ * Y
dy = Îº_Î½Î¼ * X + (Îº_Î½Î½ - 1) * X

%matplotlib inline
pyplot.figure(figsize=(10,10))
pyplot.title('Figure 4.4 (a)')
pyplot.xlabel('x(Ï)', fontsize = 14)
pyplot.ylabel('y(Ï)', fontsize = 14)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, dx, dy, linewidth = 1, density = 2)
```

### Figure 4.4

In this figure, we have a very strong internal agglomeration trend and a strong asymmetric interaction.  All fluxlines spiral approach a limit cycle, and so both populations will continue to move from section to section.

```python
N = 100                                
x_start, x_end = -1.0, 1.0
y_start, y_end = -1.0, 1.0
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X, Y = numpy.meshgrid(x, y)

Îº_Î¼Î¼ = 1.2
Îº_Î½Î½ = 1.2
Îº_Î¼Î½ = -1
Îº_Î½Î¼ = 1

dx = (Îº_Î¼Î¼ - 1) * X + Îº_Î¼Î½ * Y
dy = Îº_Î½Î¼ * X + (Îº_Î½Î½ - 1) * X

%matplotlib inline
pyplot.figure(figsize=(10,10))
pyplot.title('Figure 4.4 (a)')
pyplot.xlabel('x(Ï)', fontsize = 14)
pyplot.ylabel('y(Ï)', fontsize = 14)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, dx, dy, linewidth = 1, density = 1)
```
